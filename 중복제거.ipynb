```python
import json
import numpy as np

from PIL import Image
    
with open('unsplash_love.json') as f:
    annotations = json.load(f)
    
# resize
cnt = 0
for k, v in annotations.items():
    v = v['emotion']
    img = Image.open(k).resize((256, 256))
    img.save(k)
    cnt += 1
    if cnt % 5 == 0:
        print(cnt)
```

```python
images_dict = {}
```

```python
cnt = 0
for k, v in annotations.items():
    v = v['emotion']
    img = tuple(np.asarray(Image.open(k)).ravel().tolist())
    if img not in images_dict or len(v) > len(images_dict[img][1]):
        images_dict[img] = [k, v]
        cnt += 1
        if cnt % 5 == 0:
            print(cnt)
```

```python
cnt, len(images_dict)
```

```python
annotations_new = {}

for v in images_dict.values():
    annotations_new[v[0]] = v[1]
```

```python
annotations_new
```

```python
with open('unsplash_love_unique.json', 'w') as f:
    json.dump(annotations_new, f)
```

```python
import os

files = [os.path.join('data/images/love', f) for f in os.listdir('data/images/love')]
```

```python
for f in files:
    if f not in annotations_new:
        os.remove(f)
```
